```python
import pandas as pd
import datetime
import numpy as np
from keras.utils import np_utils
```

```python
dataset = pd.read_csv('dell_dataset.csv')
```

```python
dataset['Dates'] = [d.split('-')[2] for d in dataset['Date']]
dataset['Months'] = [d.split('-')[1] for d in dataset['Date']]
dataset['Years'] = [d.split('-')[0] for d in dataset['Date']]
```

```python
dataset = dataset[['S.No', 'Product', 'Place', 'Date', 'Festival_week', 'Dates', 'Months', 'Years', 'Sales']]
```

```python
dataset = dataset.drop(['Date'], axis=1)
dataset = dataset.drop(['S.No'], axis=1)
```

```python
dataset.sort_values(by='Years')
```

```python
data = dataset.values
```

```python
print(data[:,6])
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X_1 = LabelEncoder()
data[:, 0] = labelencoder_X_1.fit_transform(data[:, 0])

data[:, 1] = labelencoder_X_1.fit_transform(data[:, 1])

data[:, 2] = labelencoder_X_1.fit_transform(data[:, 2])
```

```python
print(len(data[:,6]))
```

```python
onehotencoder = OneHotEncoder(categorical_features = [0,1,2])
data = onehotencoder.fit_transform(data).toarray()
print(len(data[0,:]))
```

```python
X_train = data[:29216,:]
X_test = data[29216:,:]
```

```python
print(X_train.shape)
```

```python
Y_train = X_train[:,16]
X_train = X_train[:,:16]
```

```python
Y_test = X_test[:,16]
X_test = X_test[:,:16]
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
print(Y_train[40])
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier = Sequential()
```

```python
classifier.add(Dense(output_dim = 50, init = 'uniform', activation = 'relu', input_dim = 16))
```

```python
classifier.add(Dense(output_dim = 30, init = 'uniform', activation = 'relu'))
```

```python
classifier.add(Dense(output_dim = 1, init = 'uniform', activation = 'relu'))
```

```python
classifier.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
classifier.fit(X_train, Y_train, batch_size = 10, nb_epoch = 10)
```

```python
y_pred = classifier.predict(X_test)
```

```python
len(y_pred)
```

```python
mse = (np.square(y_pred - Y_test)).mean(axis=None)
```

```python
print(mse)
```
